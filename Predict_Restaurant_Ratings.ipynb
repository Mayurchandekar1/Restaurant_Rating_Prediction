```python
x=df[["Average Cost for two","Has Table booking","Has Online delivery","Price range"]]
y=df[["Aggregate rating"]]
```

```python
x
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
x["Has Table booking"]= label_encoder.fit_transform(x["Has Table booking"])
x["Has Online delivery"]= label_encoder.fit_transform(x["Has Online delivery"])
```

```python
x
```

```python
label_encoder.classes_
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
x=scaler.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
def modelresult(predictions):
    print("MSE {}".format(mean_squared_error(y_test,predictions)))
    print("MAE {}".format(np.sqrt(mean_absolute_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predictionsfromlr=lr.predict(x_test)
modelresult(predictionsfromlr)
```

```python
modelresult(predictionsfromlr)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
svr=SVR()
```

```python
param_gridsvr={"C":[0.1,0.5], "kernel" :["linear","poly"], "degree":[2] }
```

```python
gridsvr=GridSearchCV(svr,param_gridsvr)
gridsvr.fit(x_train,y_train)
```

```python
predgridsvr=gridsvr.predict(x_test)
```

```python
modelresult(predgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
parm_grid={
    "max_depth":[10,30],
    "min_samples_leaf":[1,2],
    "min_samples_split":[2,5]
}
```

```python
treemodel=DecisionTreeRegressor()
```

```python
grid_tree=GridSearchCV(estimator=treemodel,param_grid=parm_grid)
grid_tree.fit(x_train,y_train)
```

```python
treepredictions=grid_tree.predict(x_test)
```

```python
modelresult(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel=RandomForestRegressor()
```

```python
parm_gridrfr={
    "max_depth":[5],
    "n_estimators":[10]
}
```

```python
gridrfr=GridSearchCV(rfrmodel,parm_gridrfr)
gridrfr.fit(x_train,y_train)
```

```python
rfrpredictions=gridrfr.predict(x_test)
```

```python
modelresult(rfrpredictions)
```

```python
gridrfr.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn_param={"n_neighbors":[3,5,7,9]}
grid_knn=GridSearchCV(knn,knn_param)
grid_knn.fit(x_train,y_train)
```

```python
knnpredictions=grid_knn.predict(x_test)
modelresult(knnpredictions)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
ada_param={"n_estimators":[50,100,200],
           "learning_rate": [0.1,0.5,1]}
grid_ada=GridSearchCV(ada,ada_param)
grid_ada.fit(x_train,y_train)
```

```python
adapredictions=grid_ada.predict(x_test)
modelresult(adapredictions)
```
